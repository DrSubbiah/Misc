### II. Construct the transition matrix and then compute the steady-state distribution

#### Question 1 – Customer Loyalty 

A retail company tracks its customer loyalty over time. Each month, a customer can be categorized as either "Loyal" to the brand or having "Switched" to a competitor. Historical trends show that if a customer is currently loyal, there is an 80% chance they will remain loyal the following month, while the remaining 20% are likely to switch. On the other hand, if a customer has already switched, there is a 60% chance they will continue with the competitor and a 40% chance they will return to being loyal. Using this information, construct the transition probability matrix and determine the long-run proportion of customers in each category.

---

#### Question 2 – Website Navigation 

A web analytics team is studying how users move between different sections of a website: Home, Gallery, and Contact. They observe that users currently on the Home page tend to stay there with a probability of 60%, move to the Gallery page 30% of the time, and navigate to the Contact page 10% of the time. Users who are on the Gallery page remain there with a probability of 50%, while 20% go back to the Home page and 30% proceed to the Contact page. Users who are on the Contact page stay there 50% of the time, while 30% go back to the Home page and 20% visit the Gallery page. Based on this observed behavior, form the transition matrix and compute the steady-state probabilities for time spent on each page.

---

#### Question 3

A supply chain analyst monitors the daily inventory level of a warehouse, which is categorized as Full, Moderate, or Low. When the inventory is Full on a given day, it remains Full the next day with a probability of 50%, drops to Moderate with a probability of 30%, and falls to Low with a probability of 20%. If the inventory is at a Moderate level, it remains there with a probability of 40%, returns to Full with a probability of 40%, and drops to Low with a probability of 20%. When the inventory is Low, it stays Low the next day 40% of the time, improves to Moderate 30% of the time, and is restocked to Full 30% of the time. Construct the appropriate transition probability matrix and determine the long-run proportion of time the warehouse operates at each inventory level.
