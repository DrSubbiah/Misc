# <font color="maroon"> Recall - Unit II - Statistical Culture Frequentist Approach

## <font color="darkgreen"> 1. Population and Sample

In statistics, a **population** refers to the entire set of individuals or items that you are interested in studying. A **sample** is a smaller subset of the population, selected for analysis. Since it's often impractical to study the entire population, we use samples to make inferences about the population.

- **Population**: The complete set of individuals or items from which data can be collected.
- **Sample**: A smaller, representative subset of the population that is actually studied.

The goal of statistical inference is to make generalizations about the population based on information obtained from the sample.

---

## <font color="darkgreen"> 2. Frequentist Approach

**Statistical inference** is the process of using data from a sample to make conclusions about a population. In the setup of statistical inference, we typically perform two key tasks:

1. **Point Estimation**: Estimating a single value for an unknown population parameter (like the population mean or proportion).
2. **Interval Estimation**: Estimating a range of values (like a confidence interval) for the unknown population parameter.

### <font color="darkgreen"> Long-Run (Frequentist) Perspective:

- **Classical or Frequentist Inference** is based on the idea of the **long-run frequency** of events. It views probability as the limiting frequency of an event's occurrence after many repetitions or trials.

- For example, if we flip a fair coin an infinite number of times, the relative frequency of heads will converge to 50%.

- Frequentist methods typically rely on **sample data** and assume that, given infinite repetitions of the same experiment or sampling process, the estimators will converge to the true value of the population parameter.

In this setup:

- The **probability** is interpreted as the long-run frequency of an event.

- In frequentist inference, the parameter is considered a fixed but unknown quantity, while the estimator has a distribution, allowing its summaries—such as the mean and variance—to be evaluated.
  
- **Confidence intervals** are used to give a range of values where we expect the true parameter to fall with a given level of confidence based on repeated sampling.

### <font color="darkgreen"> Steps in Statistical Inference:
1. **Define the Population**: Identify the population and the parameter you want to estimate.
2. **Take a Sample**: Collect a representative sample from the population.
3. **Apply an Estimator**: Use an estimator (a statistical method) to compute an estimate of the population parameter from the sample data.
4. **Make Inferences**: Draw conclusions about the population based on the sample data and the estimator.

---

## <font color="darkgreen"> 3. Key Terms in Statistical Inference

### <font color="darkgreen"> 3.1 Estimator

An **estimator** is a statistical method or formula that is used to estimate a population parameter based on a sample. It is a rule or function that transforms sample data into an estimate of the unknown population parameter.

- **Key Concept**: Estimators are mathematical formulas or rules that produce estimates for population parameters. These estimators help bridge the gap between the sample data and the population characteristics.
- Common examples of estimators:
  - Sample mean as an estimator for population mean.
  - Sample proportion as an estimator for population proportion.

### <font color="darkgreen"> 3.2 Estimate

An **estimate** is the actual numerical value obtained by applying an estimator to a specific sample of data. It's the output or result of using an estimator on a sample.

- **Key Concept**: The estimate is the specific numerical value that approximates the population parameter. It’s the best guess about the population parameter based on the sample data.

### <font color="darkgreen"> 3.3 Likelihood Function

The **likelihood function** expresses the likelihood of observing the sample data given a set of parameter values. It is a mathematical function that quantifies how likely a particular set of parameters is, given the observed data.

- **Key Concept**: The likelihood function helps in determining which values of the parameters make the observed data most probable. This function is central to methods like Maximum Likelihood Estimation (MLE).

### <font color="darkgreen"> 3.4 Maximum Likelihood Estimates (MLE)

**Maximum Likelihood Estimation (MLE)** is a method for estimating the parameters of a statistical model by maximizing the likelihood function. The MLE is the value of the parameter(s) that maximizes the likelihood of the observed data.

- **Key Concept**: MLE finds the parameter values that make the observed data most likely by maximizing the likelihood function.

## <font color="darkgreen"> Examples

Now that we have explained the terms, let's walk through examples for each concept.

### <font color="darkgreen"> 1. Example of Population and Sample

Suppose you're interested in studying the average height of students in a school (this is your **population**). However, instead of measuring all students, you randomly select a sample of 50 students to estimate the average height for the whole school. The 50 students make up your **sample**.

---

### <font color="darkgreen"> 2. Example of Statistical Inference

Let's say you want to estimate the average score on a math test for all students in a large school district. You take a **sample** of 100 students and calculate the **sample mean**. Based on this sample mean, you then make an inference about the **population mean** score for all students in the district.

---

### <font color="darkgreen"> 3. Example of Estimator

Consider the **sample mean** as an estimator for the **population mean**. The formula for the sample mean estimator is:

$$
\hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

where:
- $ \hat{\mu} $ is the estimator for the population mean.
- $ x_i $ are the sample data points.
- $ n $ is the sample size.

For instance, if you have data from 5 students' ages: [18, 20, 22, 19, 21], the **estimator** is the formula for the sample mean.

---

### <font color="darkgreen"> 4. Example of Estimate

Using the above data of student ages ([18, 20, 22, 19, 21]), the **estimate** of the population mean is calculated by applying the **sample mean estimator**:

$$
\hat{\mu} = \frac{18 + 20 + 22 + 19 + 21}{5} = 20
$$

Thus, the **estimate** of the population mean age is 20 years.

---

### <font color="darkgreen"> 5. Example of Likelihood Function

Suppose you flip a coin 3 times and observe 2 heads and 1 tail. Let $ p $ be the probability of getting heads. The **likelihood function** $ L(p) $ for this data is:

$$
L(p) = p^2(1 - p)
$$

Here, $ p^2 $ represents the probability of getting 2 heads, and $ (1 - p) $ represents the probability of getting 1 tail. This function expresses how likely it is to observe the data (2 heads, 1 tail) for different values of $ p $.

---

### <font color="darkgreen"> 6. Example of Maximum Likelihood Estimate (MLE)

Continuing with the coin flip example, we want to find the value of $ p $ that maximizes the likelihood function $ L(p) = p^2(1 - p) $. To do this:

1. Take the derivative of the likelihood function with respect to $ p $.
2. Set the derivative equal to zero and solve for $ p $.

The MLE for $ p $ (the probability of heads) is:

$$
\hat{p} = \frac{2}{3}
$$

This is the value of $ p $ that maximizes the likelihood of observing 2 heads and 1 tail in 3 flips.

- **Estimate**: The numerical value that results from applying an estimator to a sample.

- **Likelihood Function**: A function that expresses the probability of the observed data given different values of the parameter(s).
- **Maximum Likelihood Estimate (MLE)**: The parameter value that maximizes the likelihood function, making the observed data most probable.

## <font color="darkgreen"> Bias of an Estimator

The **bias of an estimator** refers to the difference between the expected value of the estimator and the true value of the population parameter being estimated. 

Consider an **estimator** of a parameter $\theta$, which is *any* function $T=W(X_1,X_2,.......X_n)$ of a sample, a point estimator  
Then **Bias** of the estimator $T$ is defined as $\text{Bias}(T)= E(T) - \theta$.

If the expected value of the estimator is equal to the true parameter value, the estimator is considered **unbiased**. However, if there is a systematic difference, the estimator is **biased**. 

Make a note that if $E(T) = \theta$, then $T$ is called an $\textbf{unbiased estimator}$ for $\theta$    

In simple terms, bias indicates whether an estimator tends to **overestimate or underestimate** the parameter on <u>***average***</u>.

---

## <font color="darkgreen"> Variance of an Estimator

The **variance of an estimator** $T$ quantifies how much the estimator's values deviate from its expected value (the true parameter value) over repeated sampling. It is a measure of the "spread" or uncertainty of the estimator. 

For an estimator $T$ of a parameter $\theta$, the variance is given by:

$$
\text{Var}(T) = \mathbb{E}[(T - \mathbb{E}[T])^2]
$$

where:

- $\mathbb{E}[T]$ is the expected value of the estimator $T$
  
- $\mathbb{E}[(T - \mathbb{E}[T])^2]$ is the **second moment** of $T$, representing the average squared deviation from its expected value.

Alternatively, this can also be expressed as:

$$
\text{Var}(T) = \mathbb{E}[T^2] - (\mathbb{E}[T])^2
$$

where $\mathbb{E}[T^2]$ is the expected value of the square of the estimator.

- If $\text{Var}(T)$ is small, the estimator $T$ is considered **consistent** and **precise**.
- If $\text{Var}(T)$ is large, the estimator is considered **imprecise**, and additional information or refinement is needed to reduce uncertainty.

---

## <font color="darkgreen"> Mean Squared Error (MSE) of an Estimator

The **Mean Squared Error (MSE)** of an estimator $T$ is a measure of the average squared difference between the estimator and the true parameter value $\theta$. It combines both the variance of the estimator and its bias.

The MSE is defined as:

$$
\text{MSE}(T) = \mathbb{E}[(T - \theta)^2]
$$


$$E[T-\theta]^2 = E[T-E(T)+E(T)-\theta]^2$$

$$= E[(T-E(T))-(\theta-E(T))]^2$$

$$=E[(T-E(T))^2+(\theta-E(T))^2-2((T-E(T))(\theta-E(T))]$$

$$=E[T-E(T)]^2+E[\theta-E(T)]^2-2E[(T-E(T))(\theta-E(T))] \cdots \cdots (1)$$

T is a statistic, a function of $x_i$ (samples) and hence E(T) is independent of $x_i$ so that 

- $2E[(T-E(T))] = 2(\theta-E(T))E[(T-E(T))]$  $(\because$ if $a$ is a constant then, $E(aX) = aE(X))$ and
- $E(T-E(T)) = 0$ and hence
- $2E[(T-E(T))(\theta-E(T)) = 0$

$\therefore (1)$ becomes,

$$E[T-\theta]^2 = V(T)+(E(T)-\theta)^2 + 2(E(T)-\theta) E(T-E(T))$$

$$=V(T)+(E(T)-\theta)^2$$

**$$MSE_T = Var(T) + (Bias(T))^2$$**

where:

- $\text{Var}(T)$ is the variance of the estimator $T$.
- $(\mathbb{E}[T] - \theta)$ is the bias of the estimator, which measures the difference between the expected value of $T$ and the true parameter $\theta$.

$$ \therefore
\text{MSE}(T) = \text{Var}(T) + \text{Bias}^2
$$

The MSE provides a comprehensive assessment of an estimator’s accuracy, as it accounts for both variability (variance) and systematic error (bias). A smaller MSE indicates a more accurate estimator.

---

## <font color="darkgreen"> Bias-Variance Tradeoff

The **Bias-Variance Tradeoff** is a key concept when estimating a parameter $\theta$. It describes the relationship between two sources of error—**bias** and **variance**—which both contribute to the total error in an estimator.

When estimating a parameter $\theta$ using an estimator $T$, the total error can be decomposed into:

1. **Bias**: The bias of an estimator measures the difference between the expected value of the estimator and the true parameter value $\theta$. High bias means the estimator consistently overestimates or underestimates the true value.

   $$
   \text{Bias}(T) = \mathbb{E}[T] - \theta
   $$

2. **Variance**: Variance represents the variability of the estimator when applied to different data samples. High variance means the estimator's value fluctuates significantly with changes in the sample, leading to overfitting.

   $$
   \text{Var}(T) = \mathbb{E}[(T - \mathbb{E}[T])^2]
   $$

The **Bias-Variance Tradeoff** describes how these two components interact. As we make the model more flexible (for example, using more complex estimators), bias tends to decrease, but variance increases. Conversely, using simpler models reduces variance but typically increases bias.

The **Mean Squared Error (MSE)** of an estimator can be expressed as the sum of the squared bias and the variance:

$$
\text{MSE}(T) = \text{Bias}^2(T) + \text{Var}(T)
$$

- **High bias** leads to **underfitting**, where the model is too simplistic and does not capture the true parameter well.
- **High variance** leads to **overfitting**, where the model is too sensitive to variations in the data, capturing noise rather than the underlying signal.

The goal is to find a balance between bias and variance to minimize the total error. This tradeoff is central to choosing the appropriate complexity of the model or estimator when estimating a parameter.

# <font color="maroon"> Model Evaluation Metrics for Regression Models

## <font color="darkgreen">  R-squared, Adjusted R-squared, and Mean Squared Error (MSE)

Given the observed values of the dependent variable, $Y_{\text{obs}}$, and the predicted values, $Y_{\text{fit}}$, we can evaluate the performance of a regression model using several important metrics. These include **R-squared**, **Adjusted R-squared**, and **Mean Squared Error (MSE)**.

---

## <font color="darkgreen"> 1. **R-squared ($R^2$)**

R-squared, also known as the **coefficient of determination**, measures the proportion of the variance in the dependent variable that is explained by the independent variables in the model. It is a measure of the goodness of fit of the model.

### <font color="darkblue"> Formula:

$
R^2 = 1 - \frac{\sum_{i=1}^{n} (Y_{\text{obs}, i} - Y_{\text{fit}, i})^2}{\sum_{i=1}^{n} (Y_{\text{obs}, i} - \bar{Y}_{\text{obs}})^2}
$

Where:
- $Y_{\text{obs}, i}$ is the observed value of the dependent variable for observation $i$.
- $Y_{\text{fit}, i}$ is the predicted value of the dependent variable for observation $i$.
- $\bar{Y}_{\text{obs}}$ is the mean of the observed values.
- $n$ is the number of observations.

### <font color="darkblue"> Interpretation:
- **$R^2 = 1$**: The model perfectly explains all the variability of the dependent variable.
- **$R^2 = 0$**: The model explains none of the variability (i.e., the model is no better than the mean of the observed values).
- **$R^2 < 0$**: The model performs worse than a model that simply predicts the mean of the observed values (rare, but can happen in poorly fitting models).

---

## <font color="darkgreen"> 2. **Adjusted R-squared ($\bar{R}^2$)**

While **R-squared** gives us an idea of how well the model fits the data, it has limitations, especially when comparing models with different numbers of predictors. **Adjusted R-squared** accounts for the number of predictors in the model and provides a more accurate measure when comparing models with different numbers of independent variables.

### <font color="darkblue"> Formula:

$
R_{adj}^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - p - 1} \right)
$

Where:
- $R^2$ is the R-squared value.
- $n$ is the number of observations.
- $p$ is the number of independent variables (predictors).

### <font color="darkblue"> Interpretation:

- **Adjusted R-squared** adjusts for the complexity of the model. It will decrease if irrelevant predictors are added to the model.
- **$R_{adj}^2$** can be negative if the model is a poor fit and performs worse than a model predicting the mean value of the dependent variable.

---

## <font color="darkgreen"> 3. **Mean Squared Error (MSE)**

The **Mean Squared Error (MSE)** measures the average of the squared differences between the observed and predicted values. It is a common measure of model performance in regression problems, indicating how well the model's predictions match the actual data.

### <font color="darkblue"> Formula:

$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (Y_{\text{obs}, i} - Y_{\text{fit}, i})^2
$

Where:
- $Y_{\text{obs}, i}$ is the observed value of the dependent variable for observation $i$.
- $Y_{\text{fit}, i}$ is the predicted value of the dependent variable for observation $i$.
- $n$ is the number of observations.

### <font color="darkblue"> Interpretation:

- **Lower MSE**: Indicates better model performance as the difference between the predicted and actual values is smaller.
- **Higher MSE**: Indicates worse model performance, as the prediction errors are larger.

---


### <font color="darkblue"> Key Takeaways:

- **R-squared** provides a measure of how well the model explains the variance in the observed data, but it can be misleading when comparing models with different numbers of predictors.
- **Adjusted R-squared** is a more accurate measure for model comparison as it accounts for the number of predictors and penalizes the addition of irrelevant predictors.
- **Mean Squared Error (MSE)** gives a concrete measure of prediction accuracy by averaging the squared differences between observed and predicted values. Lower MSE values indicate better model performance.


### <font color="darkblue"> Explanation:

1. **R-squared** ($R^2$) indicates how much of the variance in the dependent variable is explained by the model.

2. **Adjusted R-squared** ($\bar{R}^2$) adjusts for the number of predictors in the model, providing a better measure when comparing models with different numbers of variables.

3. **Mean Squared Error (MSE)** provides a numerical estimate of how far off the predictions are, and is often used to minimize errors in the model.

---

**This Python code provides a way to calculate these three measures**

````python
import numpy as np

# Example observed and predicted values
Y_obs = np.array([3, -0.5, 2, 7, 4.2, 6.8])  # Actual observed values
Y_fit = np.array([2.8, 0.0, 2.1, 7.5, 4.0, 6.9])  # Predicted values

# Number of observations
n = len(Y_obs)

# R-squared (R²) calculation
ss_total = np.sum((Y_obs - np.mean(Y_obs)) ** 2)  # Total sum of squares
ss_residual = np.sum((Y_obs - Y_fit) ** 2)  # Residual sum of squares
r_squared = 1 - (ss_residual / ss_total)

# Adjusted R-squared calculation
p = 1  # Number of predictors (features) in the model, set to 1 for simple linear regression
adjusted_r_squared = 1 - ((1 - r_squared) * (n - 1)) / (n - p - 1)

# Mean Squared Error (MSE) calculation (from scratch)
mse = np.sum((Y_obs - Y_fit) ** 2) / n  # Mean squared error (MSE)

# Output results
print(f"R-squared (R-squared): {r_squared:.4f}")
print(f"Adjusted R-squared: {adjusted_r_squared:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
````

# <font color="maroon"> Confusion Matrix Metrics for Classification Models

Given a **confusion matrix** with the following components:

|                  | Predicted Positive (P) | Predicted Negative (N) |
|------------------|------------------------|------------------------|
| **Actual Positive (P)** | True Positive (TP)      | False Negative (FN)     |
| **Actual Negative (N)** | False Positive (FP)     | True Negative (TN)      |

The most common evaluation metrics derived from the confusion matrix are as follows:

---

## <font color="darkgreen"> 1. **Accuracy**
Accuracy is the proportion of correct predictions (both positive and negative).

$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$

---

## <font color="darkgreen"> 2. **Precision** (Positive Predictive Value)
Precision is the proportion of predicted positives that are actually positive.

$
\text{Precision} = \frac{TP}{TP + FP}
$

---

## <font color="darkgreen"> 3. **Recall** (Sensitivity or True Positive Rate)
Recall is the proportion of actual positives that are correctly identified.

$
\text{Recall} = \frac{TP}{TP + FN}
$

---

## <font color="darkgreen"> 4. **Specificity** (True Negative Rate)
Specificity is the proportion of actual negatives that are correctly identified.

$
\text{Specificity} = \frac{TN}{TN + FP}
$

---

## <font color="darkgreen"> 5. **F1 Score** (Harmonic Mean of Precision and Recall)
The F1 Score is the harmonic mean of Precision and Recall, and is particularly useful when the classes are imbalanced.

$
\text{F1 Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}} = 2 \cdot \frac{TP}{2TP + FP + FN}
$

---

## <font color="darkgreen"> 6. **False Positive Rate** (FPR)
The False Positive Rate is the proportion of actual negatives that were incorrectly classified as positives.

$
\text{False Positive Rate (FPR)} = \frac{FP}{FP + TN}
$

---

## <font color="darkgreen"> 7. **False Negative Rate** (FNR)
The False Negative Rate is the proportion of actual positives that were incorrectly classified as negatives.

$
\text{False Negative Rate (FNR)} = \frac{FN}{FN + TP}
$

---

## <font color="darkgreen"> 8. **Negative Predictive Value (NPV)**
The NPV measures the proportion of negative predictions that were actually correct.

$
\text{NPV} = \frac{TN}{TN + FN}
$

---

## <font color="darkgreen"> 9. **Area Under the Receiver Operating Characteristic Curve** (AUC-ROC)
- The **ROC curve** plots the **True Positive Rate (Recall)** against the **False Positive Rate (FPR)**.
- The **AUC (Area Under the Curve)** measures the ability of the model to distinguish between classes, ranging from 0 to 1:
  - **AUC = 1**: Perfect model.
  - **AUC = 0.5**: Random classifier.
  - **AUC < 0.5**: Worse than random (indicating poor model performance).

---

## <font color="darkgreen"> Summary of Metrics:

| Metric                      | Formula                                                          |
|-----------------------------|------------------------------------------------------------------|
| **Accuracy**                | $\frac{TP + TN}{TP + TN + FP + FN}$                             |
| **Precision**               | $\frac{TP}{TP + FP}$                                            |
| **Recall**                  | $\frac{TP}{TP + FN}$                                            |
| **Specificity**             | $\frac{TN}{TN + FP}$                                            |
| **F1 Score**                | $2 \cdot \frac{TP}{2TP + FP + FN}$                              |
| **False Positive Rate (FPR)** | $\frac{FP}{FP + TN}$                                           |
| **False Negative Rate (FNR)** | $\frac{FN}{FN + TP}$                                           |
| **Negative Predictive Value (NPV)** | $\frac{TN}{TN + FN}$                                        |
| **AUC (Area Under ROC Curve)** | Plots of TPR vs. FPR                                            |

---

**This Python code provides a way to calculate these metrics that give a comprehensive view of model's performance** 

````python
import numpy as np

# Confusion Matrix values
TP = 100  # True Positives
TN = 200  # True Negatives
FP = 5    # False Positives
FN = 10   # False Negatives

# Accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)

# Precision
precision = TP / (TP + FP)

# Recall
recall = TP / (TP + FN)

# Specificity
specificity = TN / (TN + FP)

# F1 Score
f1_score = 2 * (precision * recall) / (precision + recall)

# False Positive Rate
fpr = FP / (FP + TN)

# False Negative Rate
fnr = FN / (FN + TP)

# Negative Predictive Value
npv = TN / (TN + FN)

# Print results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"F1 Score: {f1_score:.4f}")
print(f"False Positive Rate: {fpr:.4f}")
print(f"False Negative Rate: {fnr:.4f}")
print(f"Negative Predictive Value (NPV): {npv:.4f}")
````

   $$
   R^2_{adj} = 1 - \left( \frac{5-1}{5-2} \right) (1 - 0.872) = 1 - \left( \frac{4}{3} \right) (0.128) = 1 - 0.1707 = 0.8293
   $$

3. **Mean Squared Error (MSE):**
   $$
   MSE = \frac{0.31}{5} = 0.062
   $$

---

### <font color="darkblue"> Final Answer:

- **R-squared ($R^2$):** 0.872
- **Adjusted R-squared ($R^2_{adj}$):** 0.8293
- **Mean Squared Error (MSE):** 0.062

## <font color="red"> Example Problem with Confusion matrix:

A spam filter is tested on a dataset of 100 emails. Out of the 100 emails, 50 were correctly identified as spam, 10 emails that were actually spam were missed and labelled as not spam, 5 legitimate emails were incorrectly identified as spam, and 35 legitimate emails were correctly identified as not spam. Using this data, calculate the at least three performance metrics for the spam filter
Using this data, calculate at least three performance metrics for the spam filter.

---

### <font color="darkblue">Step 1: Confusion Matrix

From the problem statement, 

- **True Positive (TP)** = 50 (correctly identified spam emails)
- **False Positive (FP)** = 5 (legitimate emails incorrectly identified as spam)
- **True Negative (TN)** = 35 (correctly identified legitimate emails as not spam)
- **False Negative (FN)** = 10 (spam emails missed and labeled as not spam)

Thus, the confusion matrix is:

|                      | Predicted: Spam | Predicted: Not Spam |
|----------------------|-----------------|---------------------|
| **Actual: Spam**     | 50 (TP)         | 10 (FN)             |
| **Actual: Not Spam** | 5 (FP)          | 35 (TN)             |

---

### <font color="darkblue"> Step 2: Performance Metrics

We can calculate the following performance metrics:

1. **Accuracy**: The proportion of correctly classified emails (both spam and not spam).

   $$
   \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
   $$

   Substituting the values:

   $$
   \text{Accuracy} = \frac{50 + 35}{50 + 35 + 5 + 10} = \frac{85}{100} = 0.85
   $$

   So, the **Accuracy** is **85%**.

2. **Precision**: The proportion of predicted spam emails that are actually spam.

   $$
   \text{Precision} = \frac{TP}{TP + FP}
   $$

   Substituting the values:

   $$
   \text{Precision} = \frac{50}{50 + 5} = \frac{50}{55} = 0.909
   $$

   So, the **Precision** is **90.9%**.

3. **Recall** (Sensitivity or True Positive Rate): The proportion of actual spam emails that were correctly identified as spam.

   $$
   \text{Recall} = \frac{TP}{TP + FN}
   $$

   Substituting the values:

   $$
   \text{Recall} = \frac{50}{50 + 10} = \frac{50}{60} = 0.833
   $$

   So, the **Recall** is **83.3%**.

4. **F1-Score**: The harmonic mean of Precision and Recall.

   $$
   \text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
   $$

   Substituting the values:

   $$
   \text{F1-Score} = 2 \times \frac{0.909 \times 0.833}{0.909 + 0.833} = 2 \times \frac{0.758}{1.742} = 0.868
   $$

   So, the **F1-Score** is **86.8%**.

---

### <font color="darkblue"> Final Results:

- **Accuracy**: 85%
- **Precision**: 90.9%
- **Recall**: 83.3%
- **F1-Score**: 86.8%

# <font color="maroon"> Bootstrap, Bagging, and Cross Validation

### <font color="darkgreen"> 1. **Bootstrap**

Bootstrap is a statistical resampling technique used to estimate the sampling distribution of an estimator by resampling with replacement from the observed data. It is commonly used to assess the performance and variability of models when the original data is limited or when we want to estimate the uncertainty associated with a given estimator.

In Bootstrap:

- **Resampling with Replacement**: A dataset is sampled multiple times to create new samples, each of the same size as the original dataset. Some data points may appear multiple times in a sample, while others may not appear at all.

- **Bootstrap Samples**: After resampling, a number of new datasets (bootstrap samples) are created. Each of these datasets is used to train the model, and the predictions are compared to the original data to assess model performance.

- **Estimate of Variability**: Bootstrap allows the calculation of standard errors or confidence intervals for model parameters by evaluating how much estimates vary across different bootstrap samples.

**Use in Model Evaluation**:
Bootstrap helps assess the variability in model performance, which can be useful for estimating the uncertainty of model predictions, and is particularly useful when traditional parametric assumptions are difficult to satisfy.

---

### <font color="darkgreen"> 2. **Bagging (Bootstrap Aggregating)**

Bagging is an ensemble learning technique that improves the accuracy and robustness of machine learning models by combining the predictions of multiple models. The key idea is to reduce variance and avoid overfitting by averaging over multiple models trained on different subsets of the data.

In Bagging:

- **Bootstrap Sampling**: Multiple bootstrap samples are drawn from the training data. Each sample is used to train a separate model (e.g., decision trees).

- **Model Aggregation**: Once all the models are trained, their predictions are combined. For regression tasks, the predictions are averaged; for classification tasks, a majority vote or probability averaging is used.

- **Parallelization**: Since each model is trained independently on a different subset of the data, bagging can be parallelized easily, making it computationally efficient.

**Key Benefits of Bagging**:

- **Reduction of Variance**: By averaging the predictions of multiple models, bagging helps to smooth out fluctuations caused by overfitting or noise in the training data.

- **Improved Accuracy**: By aggregating the outputs of several models, bagging typically improves predictive accuracy compared to a single model trained on the entire dataset.

- **Robustness**: Bagging is particularly useful for high-variance models like decision trees, where the variance can be reduced significantly.

---

### <font color="darkgreen"> 3. **Cross Validation**

Cross-validation is a model evaluation technique used to assess how a model generalizes to an independent dataset, typically by dividing the data into multiple subsets and using each subset as a testing set while the remaining data is used for training. This helps to evaluate the performance of a model more robustly and mitigates the risk of overfitting.

**Types of Cross Validation**:

- **K-Fold Cross Validation**: In this approach, the data is split into \( k \) equal-sized folds (subsets). The model is trained on \( k-1 \) folds and tested on the remaining fold. This process is repeated \( k \) times, each time using a different fold as the test set. The overall performance is averaged across all \( k \) iterations.

- **Leave-One-Out Cross Validation (LOOCV)**: This is an extreme case of k-fold cross-validation where \( k \) equals the number of data points. Each individual data point is used as a test set while the remaining data is used to train the model. LOOCV is computationally expensive but can provide an unbiased estimate of model performance, especially for small datasets.

- **Stratified Cross Validation**: In classification problems with imbalanced classes, stratified cross-validation ensures that each fold contains approximately the same proportion of each class as the entire dataset. This helps avoid issues where one fold may not be representative of the overall class distribution.

**Use in Model Evaluation**:
Cross-validation helps ensure that a model is not overfitting to the training data by testing it on multiple subsets. It provides a more reliable estimate of how the model will perform on unseen data.

---

### <font color="darkgreen"> How These Techniques Help in Finding MSE and Other Performance Metrics

- **Bootstrap**: By generating multiple samples with replacement, you can calculate the **Mean Squared Error (MSE)** for each sample and then compute an average MSE across all bootstrap samples. This provides an estimate of the variability in MSE, helping to understand how much MSE fluctuates depending on the data sampled.
  
- **Bagging**: Since bagging generates multiple models trained on different bootstrap samples, you can calculate the **MSE** for each model's predictions on the test set. The final prediction is an average (for regression) or majority vote (for classification), which generally results in a lower variance and improved MSE compared to a single model.
  
- **Cross Validation**: Cross-validation provides a way to evaluate the MSE (or other performance metrics) by repeatedly training and testing the model on different subsets of the data. By averaging the MSE across all folds, you can get a more accurate estimate of model performance and mitigate the risk of overfitting to a particular training set. Cross-validation can also help in tuning hyperparameters to minimize MSE.

Each of these methods contributes to improving model performance and gives insights into the model's ability to generalize to unseen data. They are essential techniques for assessing model accuracy and robustness, particularly when computing performance metrics like **MSE**.

## <font color="darkgreen"> Quick Example: Bootstrap, Bagging, and Cross Validation

Let’s assume we have a small dataset of 6 data points for a regression problem.

#### Dataset:
- **X**: [1, 2, 3, 4, 5, 6]
- **Y**: [12, 21, 34, 43, 56, 65]  (True values)

We want to evaluate the performance of a regression model using **Bootstrap**, **Bagging**, and **Cross Validation** 
---

#### 1. **Bootstrap Example**:
- Randomly sample with replacement from the dataset to create several bootstrap samples.
- **Fit** the model on each bootstrap sample using the entire data.
- Calculate the **MSE** for each bootstrap sample.
- Average the MSE across all bootstrap samples to estimate model performance.

---

#### 2. **Bagging Example**:
- Create 3 bootstrap samples from the dataset.
- **Fit** 3 separate models on each bootstrap sample.
- For each data point, take the average of the predictions from all 3 models.
- Calculate the final **MSE** across all predictions from the ensemble of models.

---

#### 3. **Cross Validation Example**:
- Create 3 folds (or more, but for simplicity, we choose 3).
- For each fold, **fit** the model using the entire data (as in "no splitting").
- After **fitting**, calculate the **MSE** using each fold's predictions (by using the full dataset in each iteration).
- Average the MSE across all the folds to get an estimate of performance.

---

Each of these methods gives us an estimate of model performance (MSE) by using **fitting on the entire dataset** and repeating the process, allowing us to understand variability in the model's predictions.

---

### Python Code Example:

```
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample
from sklearn.metrics import mean_squared_error

# Dataset
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
Y = np.array([12, 21, 34, 43, 56, 65])

# 1. Bootstrap Example
def bootstrap_mse(X, Y, n_iterations=100):
    mse_values = []
    for _ in range(n_iterations):
        X_resample, Y_resample = resample(X, Y, replace=True, n_samples=len(X))
        model = LinearRegression().fit(X_resample, Y_resample)
        Y_pred = model.predict(X)
        mse_values.append(mean_squared_error(Y, Y_pred))
    return np.mean(mse_values)

bootstrap_mse_value = bootstrap_mse(X, Y)

# 2. Bagging Example
def bagging_mse(X, Y, n_models=3):
    predictions = np.zeros_like(Y)
    for _ in range(n_models):
        X_resample, Y_resample = resample(X, Y, replace=True, n_samples=len(X))
        model = LinearRegression().fit(X_resample, Y_resample)
        predictions += model.predict(X)
    predictions /= n_models
    return mean_squared_error(Y, predictions)

bagging_mse_value = bagging_mse(X, Y)

# 3. Cross-Validation Example
def cross_validation_mse(X, Y, n_splits=3):
    mse_values = []
    fold_size = len(X) // n_splits
    for i in range(n_splits):
        # Select the test fold, the rest is used for fitting
        test_idx = list(range(i * fold_size, (i + 1) * fold_size))
        X_test, Y_test = X[test_idx], Y[test_idx]
        
        # Fit on the rest
        X_train = np.delete(X, test_idx, axis=0)
        Y_train = np.delete(Y, test_idx)
        
        model = LinearRegression().fit(X_train, Y_train)
        Y_pred = model.predict(X_test)
        mse_values.append(mean_squared_error(Y_test, Y_pred))
    return np.mean(mse_values)

cv_mse_value = cross_validation_mse(X, Y)

print(f"Bootstrap MSE: {bootstrap_mse_value}")
print(f"Bagging MSE: {bagging_mse_value}")
print(f"Cross-Validation MSE: {cv_mse_value}")


````python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold  # Correctly importing KFold
````

````python
# Dataset
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
Y = np.array([12, 21, 34, 43, 56, 65])

# 1. Bootstrap Example
def bootstrap_mse(X, Y, n_iterations=100):
    mse_values = []
    for _ in range(n_iterations):
        X_resample, Y_resample = resample(X, Y, replace=True, n_samples=len(X))
        model = LinearRegression().fit(X_resample, Y_resample)
        Y_pred = model.predict(X)
        mse_values.append(mean_squared_error(Y, Y_pred))
    return mse_values

bootstrap_mse_values = bootstrap_mse(X, Y, n_iterations=100)

# 2. Bagging Example
def bagging_mse(X, Y, n_models=3, n_iterations=100):
    mse_values = []
    for _ in range(n_iterations):
        predictions = np.zeros_like(Y, dtype=float)  # Initialize predictions as float64
        for _ in range(n_models):
            X_resample, Y_resample = resample(X, Y, replace=True, n_samples=len(X))
            model = LinearRegression().fit(X_resample, Y_resample)
            predictions += model.predict(X)
        predictions /= n_models
        mse_values.append(mean_squared_error(Y, predictions))
    return mse_values

bagging_mse_values = bagging_mse(X, Y, n_models=3, n_iterations=100)

# 3. Cross-Validation Example
def cross_validation_mse(X, Y, n_splits=3, n_iterations=100):
    mse_values = []
    for _ in range(n_iterations):
        kf = KFold(n_splits=n_splits)
        fold_mse_values = []
        for train_idx, test_idx in kf.split(X):
            X_train, X_test = X[train_idx], X[test_idx]
            Y_train, Y_test = Y[train_idx], Y[test_idx]

            model = LinearRegression().fit(X_train, Y_train)
            Y_pred = model.predict(X_test)
            fold_mse_values.append(mean_squared_error(Y_test, Y_pred))
        mse_values.append(np.mean(fold_mse_values))
    return mse_values

cv_mse_values = cross_validation_mse(X, Y, n_splits=3, n_iterations=100)
````

````python
print("Bootstrap MSE:")
print(np.mean(bootstrap_mse_values))

print()
print("Bagging MSE:")
print(np.mean(bagging_mse_values))

print()
print("Cross-Validation MSE:")
print(np.mean(cv_mse_values))
````
