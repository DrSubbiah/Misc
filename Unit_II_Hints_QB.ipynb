## **MLE for the Mean parameter $\mu$ of Normal Distribution**

Let $X_1,X_2,X_3,\cdots\cdots,X_n$ be random variables (identical but independent) following Normal distribution with parameters $(\mu, \, \sigma^2)$

Given that the variance $\sigma^2=9$

So only parameter to estimate is $\mu$


## **Likelihood**

Normal likelihood with unknown parameter $\mu$ and known $\sigma^2$ is given by:

$$P(\mu | \text{X}) = \prod_{i=1}^n {\frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(X_i-\mu)^2}{2\sigma^2}} }$$

$$P(\mu | \text{X}) = k \,\, e^{-\frac{\Sigma_{i=1}^n(X_i-\mu)^2}{2\sigma^2}}$$ where $k=(\frac{1}{\sigma\sqrt{2\pi}})^n,$ a constant (independent of $\mu$)

Taking natural logarithm $(ln =log_e)$,

$$l(\mu)=ln(P(\mu | \text{X}))= ln(k)-\frac{\Sigma_{i=1}^n(X_i-\mu)^2}{2\sigma^2}$$

Differentiating with respect to $\mu$,

$$\frac{dl}{d\mu} = 0 -\frac{1}{2\sigma^2}\,2\Sigma_{i=1}^n(X_i-\mu)(-1)$$

Equating the first derivative to zero,$\frac{dl}{d\mu}=0$ implies

$$ -\frac{1}{2\sigma^2}\,2\Sigma_{i=1}^n(X_i-\mu)(-1) = 0$$

$$ \implies \Sigma_{i=1}^n(X_i-\mu) = 0$$

$$ \implies \Sigma_{i=1}^n(X_i)- (n\times\mu) = 0$$

$$ \implies \Sigma_{i=1}^n(X_i) = (n\times\mu)$$

or,

$$  \hat{\mu} = \frac{1}{n} \Sigma_{i=1}^n(X_i)$$

Hence, the estimator for the mean parameter $\mu$ of Normal distribution with known variance is the sample mean $\bar{\text{X}}$

That is, $$  \hat{\mu} = \bar{\text{X}} = \frac{1}{n} \Sigma_{i=1}^n X_i$$

The random variable $(\text{X})$ can be the number of calls made by the customers and the interest is to estimate the average number of calls $\lambda$

Let us model with Poisson variable 

$X_1,...,X_n \sim \text{Poisson}(\lambda)$ where $X_i^s$ are independent, identically distributed

Then the likelihood is, 

$$f(x_i~|~\lambda) = e^{-\lambda}~ \frac{\lambda~^{x_i}}{x_i~!}$$

$$L(\lambda|X) = \prod e^{-\lambda}~ \frac{\lambda~^{x_i}}{x_i!}$$

$$=e^{-n\lambda}~ \frac{\lambda~^{\sum x_i}}{\prod x_i!}$$

Taking natural logarithm on both sides,

$$l(\lambda|X) = -n~\lambda + \sum x_i \ln \lambda + k$$

Differentiating with respect to the parameter $\lambda$ and equating to zero,

$$\frac{dl}{d\lambda} = 0$$

$$\implies -n + \sum x_i~ \frac{1}{\lambda} = 0$$

$$\implies n =\frac{\sum x_i}{\lambda}$$

or

$$\implies \lambda =\frac{\sum x_i}{n}$$

Hence, the MLE for the mean parameter of Poisson model is

$$ \hat\lambda_{ML} = \bar X$$

Given sample has $n=10$ values

$\therefore,$ estimate for the mean number of calls in the customer care auto-voicing system is

$$ \hat\lambda_{ML} = \bar X=\frac{x_1+x_2+\cdots+x_n}{n}$$

$$ \hat\lambda_{ML} = \bar X=\frac{7+9+\cdots+10}{10}=\frac{85}{10}=8.5$$

Hence, the estimated number of calls $=\,8.5$ may be approaximated to 8 or 9

This is to test whether the age group and product preference are independent or not

We use Chi-square ($\chi^2$) test of independence 

1. Calculate the total of each row $R_i$ and column $C_j$
2. Calculate the grand mean ($N=\sum_{i=1}^n R_i$, sum of all observed counts $O_{ij}$ of the table)
3. Find the expected count $E_{ij}$ for each cell using the formula $$E_{ij}=\frac{R_i \times C_j}{N}$$
4. Find the square of difference between observed and expected difference $(O_{ij}-E_{ij})^2$
5. Find the sum of the squared differences $\sum_i\sum_j (O_{ij}-E_{ij})^2$
6. Find $$\chi^2=\sum_i\sum_j\frac{(O_{ij}-E_{ij})^2}{E_{ij}}$$

$R_1=150 \, R_2=210 \,R_3= 170$

$C_1=120 \, C_2=120 \,C_3= 140 \,C_4=150$

$$\begin{pmatrix}

30 & 35 & 50 & 55
\end{pmatrix}$$

Expected Counts $E_{ij}=\frac{R_i \times C_j}{N}\,(i=1,2,3;\,j=1,2,3,4)$ are, 

$$\begin{pmatrix}

47.54717 & 47.54717 & 55.47170 & 59.43396\\
38.49057 & 38.49057 & 44.90566 & 48.11321
